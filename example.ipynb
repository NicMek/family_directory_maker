Building multiple folder structures from an excel file.

```python
import pandas as pd

df = pd.read_excel('family_list_example.xlsx')
print(df)
for i in range(df.shape[0]):
    
    Family(df["fn_dad"][i],df["ln_dad"][i],df["fn_mom"][i],df["ln_mom"][i],df.iloc[i, 4:].dropna().tolist()).build_folder(output_path)
```
